In this Notebook, we'll convert the values of the IC50 into -1 and 1

where the threshold of the IC50 for the **(5Z)-7-OXOZEAENOL** drug is the **median** of the column that represent that drug.

==> let's say the median is 0.3.

So, any value that is **greater than 0.3 is -1 (Non-effective)**

and any value that is **lesser than 0.3 is 1 (Effective)**

```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/Drug Response Prediction (DRP) Project/DRP Project/تجارب/drug_response_label_min_max_scaled.csv'
df = pd.read_csv(file_path)

# Assuming the column name for the second column is "(5Z)-7-OXOZEAENOL"
column_name = "(5Z)-7-OXOZEAENOL"

# Calculate the median of the second column
median_value = df[column_name].median()

# Apply the transformation: values above the median -> -1, values below the median -> 1
df[column_name] = df[column_name].apply(lambda x: -1 if x > median_value else 1)

# Select only the first and second columns
df_transformed = df[['ARXSPAN_ID', column_name]]

# Save the transformed dataframe to a new CSV file
output_file_path = '/content/drive/MyDrive/Drug Response Prediction (DRP) Project/DRP Project/تجارب/transformed_drug_response.csv'
df_transformed.to_csv(output_file_path, index=False)

print(f"Transformed dataset saved to {output_file_path}")
```
